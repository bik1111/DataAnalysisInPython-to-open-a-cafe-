```python
import re

df['역명'] = df['역명'].apply(lambda x: re.sub(r'\([^)]*\)','',x) if '(' in x else x)
df['역명'] = df['역명'].apply(lambda x: x+'역')

x = []
y = []
for z in df['역명']:
        
        url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(z)
        headers = {
            "Authorization": "KakaoAK " +"8eb71bfef631c75cae0d1ab43a0919b9"
        }
        
        places = requests.get(url, headers = headers).json()['documents'][0]
        x.append(places['y'])
        y.append(places['x'])
    
df['latitude'] = x
df['longitude'] = y
df2 = df.reset_index()
df2['총합산승객수'] = df2['총합산승객수'].astype(int)
        
        
```
