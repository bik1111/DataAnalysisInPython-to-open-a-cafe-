
# 8. 별점: (4)의 가공된 데이터 원본

```python
# 가설정의와 증명을 위해서 가공한 다량의 데이터들 중 일부


df.rename(columns = {'등록일자':'x'},inplace=True)
df.rename(columns = {'사용일자':'노선명','노선명':'역명','역명':'승차총승객수','승차총승객수':'하차총승객수','하차총승객수':'등록일자'},inplace=True)
df.drop(['x'],axis=1, inplace = True)
df['총합산승객수'] = df['승차총승객수'] + df['하차총승객수']
df.drop(['등록일자'],axis=1, inplace = True)
df= df.sort_values(by='총합산승객수', ascending=False)
df = df.reset_index(drop= True)
df = df.head(300)
df
```
