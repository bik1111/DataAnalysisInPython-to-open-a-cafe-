```python
# Map 호출

seoul_center = [37.517692, 126.989912]
my_map = folium.Map(location = seoul_center,
                    zoom_start = 11.5,
                    zoom_control = False, control_scale = True,
                    tiles = 'https://tiles.stadiamaps.com/tiles/outdoors/{z}/{x}/{y}{r}.png',
                    attr = '&copy; <a href="https://stadiamaps.com/">Stadia Maps</a>, &copy; <a href="https://openmaptiles.org/">OpenMapTiles</a> &copy; <a href="http://openstreetmap.org">OpenStreetMap</a> contributors')


for idx, rows in df.iterrows():
    
    # 20-24세 
    
    folium.Circle(
        location = [rows['위도'],rows['경도']],
        radius = rows['20~24세']/50,
        fill = True,
        color = 'yellow',
        popup = idx,
        tooltip = idx,
    ).add_to(my_map)
        
    # 25-29세
    folium.Circle(
        location = [rows['위도'],rows['경도']],
        radius = rows['25~29세']/50,
        fill = True,
        color = 'blue',
        popup = idx,
        tooltip = idx,
    ).add_to(my_map)
    
      # 30-34세
    folium.Circle(
        location = [rows['위도'],rows['경도']],
        radius = rows['30~34세']/50,
        fill = True,
        color = 'red',
        popup = idx,
        tooltip = idx,
    ).add_to(my_map)
    
      # 35-39세
    folium.Circle(
        location = [rows['위도'],rows['경도']],
        radius = rows['35~39세']/50,
        fill = True,
        color = 'purple',
        popup = idx,
        tooltip = idx,
    ).add_to(my_map)

    
my_map
```
