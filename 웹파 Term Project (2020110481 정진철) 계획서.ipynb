# 3. 분석을 위한 데이터 가공

오픈 API로 데이터를 불러오기 위해서 해당 라이브러리를 불러옵니다.

```python
import urllib.request 
import json  #JSON이란, JavaScript Object Notation의 약자로, key와 value 쌍으로 이루어진 형태의 object 이다.
import pandas as pd 
from pandas.io.json import json_normalize 
```

### 데이터 프레임에 저장하기

```python
url = "http://openapi.seoul.go.kr:8088/4a514a446c62696b313132665a4c436c/json/CardSubwayStatsNew/1/1000/20220301"
```

url 을 불러와 인코딩을 utf-8 로 전환하여 결과를 받습니다. <br>
또한 데이터 타입이 문자열이므로 json 으로 변환시켜줍니다.

```python
response = urllib.request.urlopen(url)  #url 파라미터에 해당 url 붙여넣기
json_str = response.read().decode("utf-8")
json_object = json.loads(json_str)
```

```python
json_object
```

이렇게 json_object 입력시 데이터의 모습을 확인 할 수 있습니다. <br>

좀 더 보기좋게 데이터를 정렬시키기 위해서 'CardSubwayStatsNew' 의 'row' 를 호출시키겠습니다.



```python
df=pd.json_normalize(json_object['CardSubwayStatsNew']['row'])
df
```
