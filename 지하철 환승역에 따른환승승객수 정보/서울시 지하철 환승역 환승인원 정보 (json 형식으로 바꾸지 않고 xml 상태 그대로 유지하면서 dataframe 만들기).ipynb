```python
from urllib.request import urlopen  #1
from bs4 import BeautifulSoup
```

```python
url = 'http://openapi.seoul.go.kr:8088/5850534c7662696b383178764a5844/xml/StationDayTrnsitNmpr/1/100'
result = urlopen(url)
house = BeautifulSoup(result, 'lxml-xml')
te = house.find_all('row')
```

```python
te
```

```python
datas = []

for i in range(len(te)):
    line = te[i].SN.string.strip()  
    stname = te[i].STATN_NM.string.strip()
    weekday = te[i].WKDAY.string.strip()
    saturday = te[i].SATDAY.string.strip()
    sunday = te[i].SUNDAY.string.strip()
  
    
    data = [line, stname, weekday, saturday, sunday]
    datas.append(data)  
```

```python
datas
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(datas, columns=['line', 'stname', 'weekday', 'saturday', 'sunday'])
```

```python
df
```

```python
df.info()
```

```python
df['line'] = df['line'].astype(int)  
df['weekday'] = df['weekday'].astype(int)  
df['saturday'] = df['saturday'].astype(int)
df['sunday'] = df['sunday'].astype(int)
```

```python
from scipy.stats import gaussian_kde
x = df["weekday"]
y = df["sunday"]

# gaussian kde 계산
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]

# 시각화
plt.figure(figsize=(8,5))
plt.scatter(x, y, c=z, s= 50, cmap=plt.cm.jet, edgecolor=['none'])
plt.colorbar()
plt.show()
```

```python
def plot_box(df, colname):
    box_ = plt.boxplot(df[colname], labels=[colname])
    plt.show()
    dict_ = {}
    dict_["label"] = colname
    dict_["cap_bottom"] = box_['caps'][0].get_ydata()[0]
    dict_["perc_25"] = box_['boxes'][0].get_ydata().min()
    dict_["median"] = box_['medians'][0].get_ydata()[1]
    dict_["perc_75"] = box_['boxes'][0].get_ydata().max()
    dict_["cap_top"] = box_['caps'][1].get_ydata()[0]
    return pd.DataFrame([dict_])
```

```python
plot_box(df, "total")
```
